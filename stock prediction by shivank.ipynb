Continuously compounded return monthly

```python
monthly_data['NESTLEINDCCReturns']=np.log(monthly_data['NESTLE']/monthly_data['NESTLE'].shift(1))
monthly_data['HINDUNILVRCCReturns']=np.log(monthly_data['HINDUSTAN UNILIVER']/monthly_data['HINDUSTAN UNILIVER'].shift(1))
monthly_data['bs100CCReturns']=np.log(monthly_data['BSE100']/monthly_data['BSE100'].shift(1))
```

```python
#This code calculates the logarithmic monthly compounded returns for three stock price columns ('NESTLE', 'HINDUSTAN UNILIVER', and 'BSE100') in the monthly_data dataset by taking the natural logarithm of the ratio between each month's stock price and the previous month's stock price. This transformation is commonly used to analyze and compare investment returns over time. The code essentially quantifies the relative growth rates of the mentioned stocks and the market index, providing a way to observe their performance dynamics on a logarithmic scale.
```

```python
monthly_data['NESTLEINDRetDiff']= monthly_data['NESTLEINDMonthlyReturns'] - monthly_data['NESTLEINDCCReturns']
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(figsize=(8,8),nrows=3, sharex=True, subplot_kw=dict(frameon=True)) # frameon=False removes frames
plt.subplots_adjust(hspace=0.01)
ax1.plot(monthly_data.index, monthly_data['NESTLEINDMonthlyReturns'], color='r',label="Simple returns NESTLE")
ax1.set_ylabel("Monthly")
ax1.axhline(y=0,color='black')
ax2.plot(monthly_data.index, monthly_data['NESTLEINDCCReturns'], color='b',label="Continuosly compounded returns NESTLE")
ax2.set_ylabel("Continuosly Compounded")
ax2.axhline(y=0,color='black')
ax3.plot(monthly_data.index, monthly_data['NESTLEINDRetDiff'], color='k',label="Difference")
ax3.set_ylabel("Difference")
ax3.set_xlabel("Index")
```

```python
#OBSERVATION
#The graph provides a comprehensive view of the monthly returns of NESTLE with distinct perspectives. The red line represents the simple monthly returns of NESTLE, showcasing the variation in gains and losses over time. The blue line depicts the continuously compounded returns of NESTLE, reflecting a smoothed representation of its performance. Comparing the red and blue lines, it's evident that continuous compounding reduces the impact of extreme fluctuations seen in the simple returns, leading to a more stabilized trajectory. The black line illustrates the difference between the simple returns and continuously compounded returns, emphasizing the impact of compounding on smoothing out short-term fluctuations. Overall, the graph effectively illustrates the transformation of NESTLE's monthly returns through both simple and continuously compounded calculations, highlighting the significance of compounding in understanding the long-term trends while mitigating the effects of short-term volatility.
```
